```python
sutter_footprint = pd.read_excel('exported_dataFrames/footprint_points_sutter.xlsx')

polygon = Polygon(zip(sutter_footprint['RAdeg'], sutter_footprint['DEdeg']))
minx, miny, maxx, maxy = polygon.bounds

minz = 0.1
maxz = 0.691
```

```python
#building resampling machine --  this works successfully

fourlac_resampled_v = pd.DataFrame()

RAdeg = []
DEdeg = []
z = []
cmvd = []
Voidiness = []

#creates randomly located sources with the same voidiness distribution 
no_samples = 0
while no_samples < 5: #only do this a few times becasue with too many sources the significance becomes infinite
    fourlac_temp = pd.DataFrame()
    fourlac_temp['Voidiness'] = fourlac_bll.Voidiness #creates 224 sources at a time

    RAdeg_temp = []
    DEdeg_temp = []
    z_temp = []
    filled = 0
    attempts = 0
    max_attempts = 1000  # safety in case it's small and hard to hit

    while (filled < len(fourlac_bll)) & (attempts < max_attempts):
        # Sample random points in the bounding box
        x_rand = np.random.uniform(minx, maxx)
        y_rand = np.random.uniform(miny, maxy)
        z_rand = np.random.uniform(minz, maxz)
        point = Point(x_rand, y_rand, z_rand)
        if polygon.contains(point):
            RAdeg_temp.append(x_rand)
            DEdeg_temp.append(y_rand)
            z_temp.append(z_rand)
            filled += 1
        attempts += 1
    
    cmvd_temp = cosmo.comoving_distance(z_temp).value # Comoving distance to void center

    fourlac_temp['RAdeg'] = RAdeg_temp
    fourlac_temp['DEdeg'] = DEdeg_temp
    fourlac_temp['z'] = z_temp
    fourlac_temp['cmvd_Mpc'] = cmvd_temp
    
    fourlac_resampled_v = pd.concat([fourlac_resampled_v, fourlac_temp], ignore_index = True)
    no_samples += 1

```

```python
#getting Fermi threshold values
filename = "detthresh_P8R3_12years_PL22.fits" #sent by Olivier

# Open the FITS file
with fits.open(filename) as hdul:
    # 2D array in Galactic coordinates
    data = hdul[0].data
    header = hdul[0].header
    
wcs_Fermi = WCS(header)

#converting 4LAC BL Lac locations to pixels on the fermi threshold map for comparison
four_lac = pd.read_excel('exported_dataFrames/4lacsutter_bll.xlsx')
FLac_celestial = SkyCoord(four_lac.RAdeg.values* u.degree, four_lac.DEdeg.values* u.degree, frame='icrs')
# Transform to galactic coordinates
FLac_galactic = FLac_celestial.transform_to('galactic')
#convert to pixels
pix_x, pix_y = wcs_Fermi.wcs_world2pix(FLac_galactic.l, FLac_galactic.b, 1)


int_pix_x = np.round(pix_x).astype(int)
int_pix_y = np.round(pix_y).astype(int)


sensitivity_4lac =[] #creating 
for i in range(len(pix_x)):
    sensitivity_4lac.append(data[int_pix_y[i],int_pix_x[i]])
```

```python
#from Olivier

def fermi_threshold(voidiness):
    a = 6.4e-14
    b = 1.1e-12
    return a*voidiness + b
```
